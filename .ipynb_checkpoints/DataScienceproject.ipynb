```python
import matplotlib.pyplot as plt
import pandas as pd
AB = pd.read_csv("AB.csv")
AB.head()
plt.style.use('Solarize_Light2')
```

```python
boroughs = AB.groupby('neighbourhood_group').id.count()
plt.figure(figsize = (100,50))
plt.rc('xtick', labelsize= 75)
plt.rc('ytick', labelsize= 75)
plt.xlabel('Boroughs', fontsize = 75)
plt.title('AirBnBs per borough', fontsize = 75)
boroughs.plot(kind = 'bar')
```

Based on the graphed data, the highest/lowest points on the graph corresponds to the boroughs of least and most interest. A place like Staten Island and the Bronx has significantly less interest than a place like Manhattan and Brooklyn, which are major Tourist attractions within New York City. 

```python
RoomType = AB.groupby(['room_type', 'neighbourhood_group']).id.count()
plt.rc('xtick', labelsize= 10)
plt.rc('ytick', labelsize= 10)
RoomType.unstack().plot(kind = 'bar')
plt.xlabel('AirBnB Rooming types & borough')
plt.title('Variety of AirBnB types around NYC')
```

This graph clarly shows a correlation between the popular forms of AirBnBs which also complement the pervious graph. Due to the fact people have less interest in AirBnbs such as shared rooms, the number of those rooming types are significantly lower; however, this also can be used to show what type of boroughs you could expect a specific rooming type in.

```python
BoroughAvg = AB.groupby(['neighbourhood_group']).price.mean()
plt.figure(figsize = (100,50))
plt.rc('xtick', labelsize= 75)
plt.rc('ytick', labelsize= 75)
plt.title('Average AirBnB cost per borough', fontsize= 75)
BoroughAvg.plot(kind = 'bar')

```

When analyzing the graph, it is crucial to consider the significance of the pricing gap between Manhattan and every other borough due to the fact that a borough such as Manhattan is a major tourist attraction while on the contrary, a place like the Bronx isn't really visited by any tourist.

```python
AvgAvailability = AB.groupby(['neighbourhood_group']).availability_365.mean()
plt.figure(figsize = (100,50))
plt.rc('xtick', labelsize= 75)
plt.rc('ytick', labelsize= 75)
plt.title('Average Availability for AirBnB per borough', fontsize = 75)
AvgAvailability.plot(kind = 'bar')
```

It is interesting to note the difference in values between places like Brooklyn and Staten Island. It is possible to assume that based on all of the data previously shown, Brooklyn might be slightly more popular than Manhattan due to having a lower average cost for Airbnbs while being close to Manhattan, allowing for a more afforable planning. Where as a place like Staten Island would clearly have more avalibility due to it being more remote than others.

*Use the code block below to open `sales.csv` with python. Then create a variable called `content` which is equal to a list of each line of the file*

````python
#opens files and relables as 'sales'
with open('sales.csv') as sales :
    #Reads the file and sets to content variable
    content = sales.readlines()
    
````

*Using the `split` function and a `for` loop calculate the total sales in a month*

*Hint: you will need a seperate variable for each month*

````python
January = 0
Febuary = 0 
March = 0

del(content[0])
for a in content :
    Lst = a.split(',')
    total = Lst[2]
    date = Lst[3]   
    Lst = date.split('/')
    if Lst[0] == '1':
        January += float(total)
    elif Lst[0] == '2':
        Febuary += float(total)
    elif Lst[0] == '3':
        March += float(total)

print(January)
print(Febuary)
print(March)
````

*Using your calculated total sales create a graph to represent this data and write a written analysis*

````python
import matplotlib.pyplot as plt

MonthsOfYear = ['January','Febuary','March']
Sales = [January,Febuary,March]

plt.plot (MonthsOfYear, Sales)
plt.title ('Value of Revenue for monthly sales')
````

The graph shown above is a representation of Revenue made due to the combanies sales per month, showing which month had the lowest and the highest revenue.

*Using the `split` function, seperate each row of data. Then append the customer's name to a list called `customers`*

````python
customers = []
for a in content:
    CustomerData = a.split(',')
    customers.append(CustomerData[0])
del (customers[0])
print(customers)
````

*Using the `Counter` from the `collections` library, and your list of customers create a graph to represent this data and write a written analysis*

````python
Customers_counter = Counter(customers)

customers = Customers_counter.keys()
arrivals = Customers_counter.values()

plt.bar(customers,arrivals)
plt.rc('xtick',  labelsize=50,)
plt.rc('ytick', lablesize=50,)
plt.figure(figsize=(100,50))
````
